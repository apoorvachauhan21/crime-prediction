```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory
```

```python
vic_rape_df = pd.read_csv("C:/MNNIT_Python/crime_data/20_Victims_of_rape.csv")
#Exploring first few rows of 'Victims of Rape' dataset
vic_rape_df.head()
```

```python
state=input()
mah_vic=vic_rape_df.loc[vic_rape_df['Area_Name']==state]
mah_vic = mah_vic [(mah_vic['Subgroup']=='Total Rape Victims')]

#Plotting age breakup of victims
```

```python
mah_vic
```

```python
mah_vic.corr()
```

```python
Y = mah_vic.Victims_of_Rape_Total.values
import numpy as np
X=mah_vic.Year[:,np.newaxis]
X
```

```python
Y
```

```python
Xmin = X.min()
print("X min",Xmin)
Xmax = X.max()
print("X max",Xmax)
Xnorm = (X -Xmin)/(Xmax-Xmin)
print(Xnorm)
Ynorm = (Y - Y.min())/(Y.max()-Y.min())
print(Ynorm)
#Put the norm values into dataframe
df2 = mah_vic
df2.Year = Xnorm
df2.Victims_of_Rape_Total = Ynorm
#Plot the graph
df2.plot()
```

```python
import sklearn.linear_model as lm
lr = lm.LinearRegression()
# Training the model using given data
# Y = b0+b1*X
#b0 = intercept
#b1 = coeff
lr.fit(X,Y)  #Training the model
#Prediction
year=int(input("Enter the year"))
noc = lr.predict(year)
print("In Year ",year," there is probability of ",int(noc)," Cases")
```

```python
import matplotlib.pyplot as plt
# Interpret : Actual vs Predicted
# plot actual marks vs hours
plt.plot(X, Y,color="red")
plt.scatter(X,Y,color="red")
plt.xlabel("Cases")
plt.ylabel("Year")
plt.title("Actual vs Predicted cases")
plt.plot(X, lr.predict(X),color="green")
plt.scatter(X, lr.predict(X),color="green")
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(Y,lr.predict(X))
print("Accuracy is ",accuracy)
#Errors: 0 - n
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(Y,lr.predict(X))
print("Mean absolute error ",mae)
```
